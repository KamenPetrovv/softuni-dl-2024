Standard ML:
- Pros:
    - If you have a clear idea about the logic(math) involved in behind and you can visualize the complex functionalities like K-Mean, Support Vector Machines, etc., then it defines the ML aspect.
    - **Work well with tabulated data**
    - ML is less complex and has a lower data volume.
- Cons:
    - Less efficient than DL as it can’t work for longer dimensions or higher amount of data.
    - Doesn't work well with unstructured data.
    - You have to manually select and extract features from raw data and assign weights to train an ML model.
     
NNs:
- Pros:
    - **The freedome they give us**
    - **We can create our own loss function (that is differentiable), hence we can solve any problem that we can define a loss function for**
    - More powerful than ML as it can easily work for larger sets of data.
    - **Works great with unstructured data (untabulated data).**
    - **Utilize GPUs better, but VRAM is often a bottleneck**
    - **Can work with huge amounts of parameters (billions) $10^{11}$**
    - Unlimited capacity
    - No feature selection/**creation** (A NN can figure out what a nose/face is on its own) (Deep learning models can self-learn using feedback from known errors.)
- Cons:
    - If you are clear about the math involved in it but don’t have idea about the features, so you break the complex functionalities into linear/lower dimension features by adding more layers, then it defines the DL aspect.
    - **Can't interpret the weights. Black box.**
    - Doesn't work well with tabulated data
    - Deep learning is more complex with a very high data volume.

## Neural Networks

It is basically a DAG (Directed acyclic graph)



They can have all shapes as we will see in the future, but for now let's look at the basic one:
#### Feedforward NN


Charcteristics:
- They have an input layer
- They have an output layer
- They have a number of hidden layers, that increase its capacity
- They have a direction (forward only), we cannot go back (becasuse we will create cycles) 
- The neurons in each layer are NOT connected

cuDNN - CUDA® Deep Neural Network

cuDNN is a GPU-accelerated library of primitives(lego building blocks) for deep neural networks. It is build on top of CUDA.

#### What is @tf.function?

[tf.function is a decorator function provided by Tensorflow 2.0 that converts regular python code to a callable Tensorflow graph function, which is usually more performant and python independent. It is used to create portable Tensorflow models.](https://www.machinelearningplus.com/deep-learning/how-use-tf-function-to-speed-up-python-code-tensorflow/)

This is why TF2.0 has the tf.function API, to give any user the option to convert a regular (eager) python code to a lazy code which is actually speed optimized.

But, why use Graphs? :
- The primary reason is, graphs allow your neural network model to be used in enviornments that don't have a Python interpreter. For example, graphs can be deployed in mobile applications or servers. This is not suitable for eagerly executed code.
- The second reason is that graphs can speed up computation time. They eliminate the need for repetitive initialisation of variables and computation on these variables.les.

Loss function for:

Different kidns of classification loss functions:
- [BinaryCrossentropy](https://keras.io/api/losses/probabilistic_losses/#binarycrossentropy-class) - The output can be 0 or 1
- [CategoricalCrossentropy](https://keras.io/api/losses/probabilistic_losses/#categoricalcrossentropy-class) - The outputs are two or more label classes. We expect labels to be provided in a one_hot representation
- [SparseCategoricalCrossentropy](https://keras.io/api/losses/probabilistic_losses/#sparsecategoricalcrossentropy-class) -The outputs are two or more label classes. We expect labels to be provided as integers

`loss = "sparse_categorical_crossentropy"` Gives us probabilities in the outputs. We have to make sure that they add up to 1.0. To do this we have to use the right activation function when defining the Output Dense layer. `Dense(number_of_classes, activation = "softmax")`

Softmax works just like sigmoid, but for multiple classes.

*If we don't give an activation function, the NN will collapse to a really slow Linear Regression.

````python
#model_tf.compile(loss= keras.losses.SparseCategoricalCrossentropy())
model_tf.compile(loss = "sparse_categorical_crossentropy", optimizer = "adam") 
#optimizer (or solver in  scikit-learn) the algo we will be using for gradient descent. More on this later.
````

````python
model_tf.fit(iris_attributes, iris_labels)
````

````python
def train(model, optimizer, criterion, X, y, num_epochs, train_losses):
    for epoch in range(num_epochs):
        optimizer.zero_grad()
        output_train = model(X) # forward
        loss_train = criterion(output_train, y)
        loss_train.backward() # backward
        optimizer.step() # weight update
        train_losses[epoch] = loss_train.item()
        if (epoch + 1) % 50 == 0:
            print(f"Epoch {epoch+1}/{num_epochs}, Loss: {loss_train.item():.4f}")
````
